# Regex Symbols Meaning

| Symbol | Meaning | Example |
|--------|---------|---------|
| `.` | Koi bhi ek character | `a.c` → abc, a1c |
| `^` | String ki starting check karta hai | `^Hello` → Hello se start |
| `$` | String ka ending check karta hai | `com$` → com par end |
| `*` | 0 ya usse zyada times | `a*` → "", a, aa, aaa |
| `+` | 1 ya usse zyada times | `a+` → a, aa, aaa |
| `?` | 0 ya 1 time (optional) | `colou?r` → color, colour |
| `{n}` | Exact n times | `\d{10}` → 10 digits |
| `{n,m}` | Minimum n aur maximum m times | `\d{2,4}` → 12, 123, 1234 |
| `[]` | Character set | `[abc]` → a ya b ya c |
| `[^ ]` | Not (exclude) | `[^0-9]` → numbers ke alawa |
| `()` | Group banata hai | `(abc)+` |
| `|` | OR condition | `cat|dog` → cat ya dog |

```python
import re # importing re modules

txt ="hello neha how are you"
x=re.search("neha",txt)
if x:
    print("yes! we have a match")
else:
    print("No Match")
```

#### Use case of re modules

### 1.Email Extract karna(re.findall)

text="contact nehakarki974@gmail.com"


```python
import re # importing re modules
text="contact nehakarki974@gmail.com nk590375@gmail.com"
email=re.findall(r'\S+@\S+',text)
print(email)
```

### 2. Phone Number Extract karna


```python
import re
text="my contact number is 9058348934"
number=re.findall(r'\d{10}',text)
print(number)        
```

### 3. Sare Numbers Extarct karna

```python
text= 'I bought 2 phone and 10 laptops for comapny uses'
numbers=re.findall(r'\d+',text)
print(numbers)
```

### 4. Special Characters Remove karna (re.sub())

```python
text = "Hello@# Python world!!"
```

```python
import re

clean_text = re.sub(r'[^a-zA-Z ]', '', text)

print(clean_text)
```

### 5. Replace Numbers with X

```python
import re

text = "My bank OTP is 123456"

result = re.sub(r'\d', 'X', text)

print(result)
```

### 6. Password Validation

```python
import re

password = "Neha1234"

pattern = r'^(?=.*[A-Z])(?=.*\d).{8,}$'

result = re.match(pattern, password)

if result:
    print("Valid Password")
else:
    print("Invalid Password")
```
